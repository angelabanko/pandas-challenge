## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
new_price = purchase_data.loc[:,["Gender", "Price"]]
group_gender = new_price.groupby(["Gender"])
count_group = group_gender.count()
female_sum= new_price.groupby("Gender")["Price"].sum()['Female']
male_sum= new_price.groupby("Gender")["Price"].sum()['Male']
other_sum= new_price.groupby("Gender")["Price"].sum()['Other / Non-Disclosed']
female_mean= round(new_price.groupby("Gender")["Price"].mean()['Female'],2)
male_mean= round(new_price.groupby("Gender")["Price"].mean()['Male'],2)
other_mean= round(new_price.groupby("Gender")["Price"].mean()['Other / Non-Disclosed'], 2)
female_ave_total = round(female_sum/females, 2)
male_ave_total = round(male_sum/males, 2)
other_ave_total = round(other_sum/other, 2)
list_count_group = count_group.values.tolist()
female_count = list_count_group[0]
male_count =list_count_group[1]
other_count = list_count_group[2]
purchase_df = pd.DataFrame({"Gender": ["Female", "Male", "Other / Non-Disclosed"],
                            "Purchase Count": [female_count[0], male_count[0], other_count[0]],
                           "Average Purchase Price": [female_mean, male_mean, other_mean],
                            "Total Purchase Value": [female_sum, male_sum, other_sum],
                            "Avg Total Purchase Per Person": [female_ave_total, male_ave_total,other_ave_total]})
purchase_df['Average Purchase Price'] = purchase_df['Average Purchase Price'].map('${:,.2f}'.format)
purchase_df['Total Purchase Value'] = purchase_df['Total Purchase Value'].map('${:,.2f}'.format)
purchase_df['Avg Total Purchase Per Person'] = purchase_df['Avg Total Purchase Per Person'].map('${:,.2f}'.format)
gender_index = purchase_df.set_index("Gender")
gender_index
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_price_df = purchase_data.loc[:,["SN", "Price"]]
sn_count = sn_price_df.groupby('SN').count()
sn_count= sn_count.rename(columns={'Price':'Purchase Count'})
sn_total = sn_price_df.groupby("SN")
total_sum = sn_total.sum()
total_sum = total_sum.rename(columns={'Price':'Total Purchase Value'})
sn_df = pd.concat([sn_count, total_sum], axis=1)
sn_df['Average Purchase Price'] = sn_df['Total Purchase Value']/sn_df['Purchase Count']
sn_df = sn_df[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
sn_df['Average Purchase Price']= sn_df['Average Purchase Price'].round(2)
sorted_sn_df = sn_df.sort_values('Total Purchase Value', ascending=False)
sorted_sn_df['Average Purchase Price'] = sn_df['Average Purchase Price'].map('${:,.2f}'.format)
sorted_sn_df['Total Purchase Value'] = sn_df['Total Purchase Value'].map('${:,.2f}'.format)

sorted_sn_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
popular_item_df = item_df.groupby(["Item ID", "Item Name"])
popular_item_sum = popular_item_df.sum()
popular_mean = popular_item_df.mean()
popular_item_sum = popular_item_sum.rename(columns={"Price":"Total Purchase Value"})
popular_mean = popular_mean.rename(columns={"Price":"Item Price"})
popular_counts = item_df.groupby(["Item ID", "Item Name"]).count()
popular_counts = popular_counts.rename(columns={"Price":"Purchase Count"})
popular_data = pd.concat([popular_counts, popular_mean, popular_item_sum], axis=1)
sorted_popular_data = popular_data.sort_values("Purchase Count", ascending=False)
sorted_popular_data["Item Price"] = popular_data["Item Price"].map('${:,.2f}'.format)
sorted_popular_data["Total Purchase Value"] = popular_data["Total Purchase Value"].map('${:,.2f}'.format)
sorted_popular_data.head()
```
